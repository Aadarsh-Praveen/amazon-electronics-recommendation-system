```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
s3_path = "s3://amazon-electronics-dataset/Raw_Dataset/Electronics_5.json"
df = pl.scan_ndjson(s3_path)
```

```python
df_nulls = df.select([
    pl.col(col).is_null().sum().alias(f"{col}_nulls")
    for col in df.schema.keys()
])

df_nulls.collect()
```
